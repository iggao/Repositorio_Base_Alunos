# AND

True + True = True  
True + False = False  
False + True = False  
False + False = False

```python
(10>20) and (20>30) # False + False
```

```python
(10<20) and (20<30) # True + True
```

```python
(20>30) and (10<20) # False = True
```

# OR

True + True = True  
True + False = True   
False + True = True   
False + False = False   

```python
(1 < 2) or (3 > 6) # True + False
```

```python
'João' == 'João' or "Python" != 'python'
```
