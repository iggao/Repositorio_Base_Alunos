# Operadores Aritimeticos

Adição (+)

```python
print(10 + 20)
print(10 + 20.5)
print(10 + 30 + 40)

print(-89 + 63) # print() é uma função de saída no python
print(23 - 17) # 23 + (-17)
print(23 + (-17) )
print(34 - 67) # 34 + (67)
```

Subtração (-)

```python
print(235-234)
```

Multiplicação (*)

```python
print(12*2)
print(2.4*3.2)
```

Divisão (/)

```python
print(1000/10)
print(129/23)
```

Divisão com resultado inteiro (//)

```python
print(129//23)
```

Multiplicação com expoente (**)

```python
# Exemplo de uso do operador ** para expoentes
# base = 2
# expoente = 3
# resultado = base ** expoente
# print(f"{base} elevado a {expoente} é igual a {resultado}")

# expoente positivo
print(2 ** 3)

# expoente negativo
print(2 ** -1)

# expoente decimal negativo
print(2 ** -1.5)

# expoente decimal positivo
print(2 ** 1.5)
```

 1. Crie um código que receba o nome do aluno, duas notas e informe a média do aluno. Obs: no seu print deve apresentar o nome do aluno, as duas notas e a média

```python

nome = input("Digite o nome do aluno: ")
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))

media = (nota1 + nota2) / 2

print(f"O(A) aluno(a) {nome} obteve as notas {nota1} e {nota2}, e a sua média é {media}.")
```

2. Crie um código que receba as medidas de um retângulo e informe o perimetro do mesmo.

```python

lado1 = float(input("Digite a medida do primeiro lado do retângulo: "))
lado2 = float(input("Digite a medida do segundo lado do retângulo: "))

perimetroretangulo = 2 * (lado1 + lado2)

print(f"O perímetro do retângulo com lados {lado1} e {lado2} é {perimetroretangulo}.")
```

3. Crie um código que calcule a raiz quadrada de um número.

```python
num = float(input("Digite um numero para calcular a raiz quadrada : "))
raizquadrada = num ** 0.5
print(f"A raiz quadrada de {num} é {raizquadrada}")
```

8) Crie um programa que solicite ao usuário a quantidade de horas trabalhadas por semana e o valor da hora de trabalho e calcule o salario mensal.

```python

horasporsemana = float(input("Digite a quantidade de horas trabalhadas por semana: "))
valorhora = float(input("Digite o valor da hora de trabalho: "))

salariomensal = horasporsemana * valorhora * 4

print(f"O salário mensal estimado é de R${salariomensal:.2f}")
```
